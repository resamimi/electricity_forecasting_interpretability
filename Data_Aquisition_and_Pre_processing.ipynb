```python
from hampel import hampel
```

```python
renergeticData = pd.read_csv("/content/drive/MyDrive/Interpretable Forecasting of Buildings Energy Consumption/code/outputs/data/renergetic/debit2_2020_2021.csv")
```

```python
renergeticData.head()
```

```python
len(renergeticData)
```

```python
fig = px.line(renergeticData, x="timestamp", y=["MW"], template = 'ggplot2')
fig.show()
```

```python
window_size=14*24

# Just outlier detection
outlier_indices = hampel(renergeticData["MW"], window_size)
print("Outlier Indices: ", outlier_indices)
print("Number of outliers: ", len(outlier_indices))
```

```python
# Outlier Imputation with rolling median
renergeticData_imputation = renergeticData.copy()
renergeticData_imputation["MW"] = hampel(renergeticData["MW"], window_size, imputation=True)
```

```python
fig = px.line(renergeticData_imputation, x="timestamp", y=["MW"], template = 'ggplot2')
fig.show()
```

```python
renergeticData_imputation.head()
```

```python
for index, row in renergeticData.iterrows():
  if index in outlier_indices:
    print('time:', renergeticData.iloc[index]['timestamp'])
    print('before', renergeticData.iloc[index]['MW'])
    print('after' , renergeticData_imputation.iloc[index]['MW'])
    print('-------------------------------------------------')
```

```python
renergeticData_imputation.to_csv("/content/drive/MyDrive/Interpretable Forecasting of Buildings Energy Consumption/code/outputs/data/renergetic/debit2_2020_2021_NoOutlier.csv")
renergeticData_imputation.to_csv("/content/drive/MyDrive/Interpretable Forecasting of Buildings Energy Consumption/code2/outputs/data/renergetic/debit2_2020_2021_NoOutlier.csv")
```
